```python
import numpy as np
import matplotlib.pyplot as plt
import h5py
```

```python
with h5py.File('BBH_dataset.hdf5', 'r') as f:
    print(f.keys())
    print(list(f.attrs))
```

```python
with h5py.File('BBH_dataset.hdf5', 'r') as f:
    data = np.array(f['h_plus'])
    mass1 = np.array(f.attrs['mass1'])
    mass2 = np.array(f.attrs['mass2'])
print(data.shape)
print(mass1.shape)
```

```python
fig, axs = plt.subplots(2, 5, figsize=(16, 6))
for ith_ax, ax in enumerate(axs.flatten()):
    ax.plot(data[ith_ax])
    ax.set_title(f'$M_1$ = {mass1[ith_ax]}, $M_2$ = {mass2[ith_ax]}')
fig.tight_layout()
plt.show()

# 
```
